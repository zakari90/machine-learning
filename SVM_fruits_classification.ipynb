```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = "/content/drive/MyDrive/hsoub ai/machine learning/apples_and_oranges.csv"
df = pd.read_csv(file_path)

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot of Weight vs. Size, colored by Class
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Weight', y='Size', hue='Class', palette='viridis', s=100)
plt.title('Weight vs. Size colored by Class')
plt.xlabel('Weight (grams)')
plt.ylabel('Size (cm)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df[['Weight','Size']]
y = df['Class']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y)

from sklearn.svm import SVC

model = SVC(kernel='linear')

model.fit(X_train.values, y_train)
```

```python
y_pred=model.predict(X_test.values)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('\nAccuracy: {:.2f}\n'.format(100*accuracy_score(y_test, y_pred)))

print('\nPrecision: {:.2f}\n'.format(100*precision_score(y_test, y_pred)))

print('\nRecall: {:.2f}\n'.format(100*recall_score(y_test, y_pred)))

print('\nF1: {:.2f}\n'.format(100*f1_score(y_test, y_pred)))
```

```python
Weight=150
Size=5.0

Classpred = model.predict([[Weight,Size]])
Classpred = le.inverse_transform(Classpred)
print(Classpred)
```
