```python
import pandas as pd

dataset = pd.read_csv("/content/drive/MyDrive/hsoub ai/machine learning/regression/iPhone.csv")
dataset.info()
```

```python
dataset.head(2)
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, 3]
```

```python
# X = dataset[["Gender","Age","Salary"]]
# Y = dataset["Purchase Iphone"]
```

```python
from sklearn.preprocessing import LabelEncoder
laberEconder_gender = LabelEncoder()
X[:,0]= laberEconder_gender.fit_transform(X[:,0])
X[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler. This object will be used to scale the features.
sc = StandardScaler()

# Fit the StandardScaler to the training data (X_train) and then transform it.
# 'fit' calculates the mean and standard deviation for each feature in X_train.
# 'transform' then applies this scaling to X_train, resulting in a new, scaled X_train.
X_train = sc.fit_transform(X_train)

# Transform the test data (X_test) using the scaling parameters learned from the training data.
# It's crucial to only 'transform' the test data, not 'fit_transform', to prevent data leakage
# from the test set and ensure consistency with how the training data was scaled.
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver="liblinear")
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn import metrics
confusion_metrics = metrics.confusion_matrix(y_test, y_pred)
accuarcy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred)
recall = metrics.recall_score(y_test, y_pred)

print("cm : " , confusion_metrics)
print("accuarcy :", accuarcy)
print("precision :", precision)
print("recall :", recall)
```
